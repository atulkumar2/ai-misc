# Partial Dependence plot

Partial Dependence Plots are powerful tools for interpreting machine learning models by visualizing the marginal effect of a feature on the predicted outcome. They help demystify black-box models by showing whether relationships between features and predictions are linear, monotonic, or more complex. However, PDPs must be interpreted with caution due to the assumption of feature independence and potential computational limitations for large datasets.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_friedman1
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.inspection import PartialDependenceDisplay
```

```python
# Creating synthetic dataset
X, y = make_friedman1(n_samples=1000, n_features=10, random_state=42)

# Train a Gradient Boosting Regressor
model = GradientBoostingRegressor(n_estimators=100, max_depth=4, random_state=42)
model.fit(X, y)
```

```python
# Plotting PDP for feature 0 and feature 1
features = [0, 1]
display = PartialDependenceDisplay.from_estimator(model, X, features, grid_resolution=50)

plt.show()
```
